```
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules 
```

```
data = pd.read_csv('/content/drive/My Drive/Python/apriori.csv') 
data.head() 
```

```
# Building the model 
frq_items = apriori(data, min_support = 0.4, use_colnames = True)

# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, min_threshold=0) 
rules = rules.sort_values(['confidence'], ascending =[False]) 
display(rules[['antecedents', 'consequents', 'support', 'confidence']])
```

```
cols = ['antecedents','consequents']
rules[cols] = rules[cols].applymap(lambda x: tuple(x))
```

```
rules.to_excel('/content/drive/My Drive/Python/apriori_res40.xlsx', index=False, columns=['antecedents', 'consequents', 'support', 'confidence'])
```
